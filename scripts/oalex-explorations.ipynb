```python
import requests
import pandas as pd
import numpy as np
```

```python
# Initialize API URL & Parameters
base_url = "https://api.openalex.org/sources"  # OpenAlex API endpoint
email = "kase@ff.zcu.cz"  # Provide your valid email address
cursor = "*"  # Initial cursor (start from the beginning)
per_page = 200  # Number of results per page

# Container to hold all extracted data
all_sources = []

# Loop to fetch all pages
while cursor:
    params = {
        "per_page": per_page,  # Maximum allowed results per page
        "cursor": cursor,  # Cursor to fetch data
        "mailto": email  # Add email for polite pooling
    }

    try:
        # Request data from API
        response = requests.get(base_url, params=params)
        response.raise_for_status()  # Raise an exception for HTTP errors
        data = response.json()
    except requests.exceptions.RequestException as e:
        break

    # Append results to all_sources
    all_sources.extend(data.get("results", []))

    # Fetch the next cursor
    cursor = data.get("meta", {}).get("next_cursor", None)
    if not cursor:  # If no next cursor, exit the loop
        print("No more results to fetch.")
        break
```

```python
sources_df = pd.DataFrame(all_sources)
sources_df.head()
```

```python
sources_df.to_parquet("../data/sources.parquet")
```

```python
sources_df = pd.read_parquet("../data/sources.parquet")
sources_df.head(5)
```

```python
sources_df["topics_lists"] = sources_df["topics"].apply(
    lambda x: [t["display_name"] for t in x] if isinstance(x, (list, np.ndarray)) else [])

sources_df["domains_list"] = sources_df["topics"].apply(
    lambda x: list(set([t["domain"]["display_name"] for t in x])) if isinstance(x, (list, np.ndarray)) else [])

sources_df["fields_list"] = sources_df["topics"].apply(
    lambda x: list(set([t["field"]["display_name"] for t in x])) if isinstance(x, (list, np.ndarray)) else [])
```
