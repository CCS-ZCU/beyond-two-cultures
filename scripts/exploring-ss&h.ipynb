```python
def oldest_year(counts_by_year):
    # Check for None or empty numpy array
    if counts_by_year is None or len(counts_by_year) == 0:
        return None  # Return None for empty or invalid input

    try:
        # Extract and collect all "year" values
        valid_years = [
            item["year"]
            for item in counts_by_year
            if isinstance(item, dict) and "year" in item and isinstance(item["year"], (int, float))
        ]
        
        # Return the minimum year if valid years exist, otherwise None
        return min(valid_years) if valid_years else None
    except Exception as e:
        # Handle unexpected errors gracefully
        print(f"Error occurred: {e}")
        return None
```

```python
import requests

# Initialize API URL & Parameters
def get_works_from_source(source_id, email = "kase@ff.zcu.cz"):
    base_url = "https://api.openalex.org/works"  # OpenAlex API endpoint for works
    cursor = "*"  # Initial cursor (start from the beginning)
    per_page = 200  # Maximum results per page

    # Container to hold all extracted data
    all_works = []

    # Loop to fetch all pages
    while cursor:
        params = {
            "filter": f"locations.source.id:{source_id}",  # Filter works by source ID
            "per_page": per_page,  # Limit results per page
            "cursor": cursor,  # Cursor for pagination
            "mailto": email  # Add email for polite pooling
        }

        try:
            # Request data from API
            response = requests.get(base_url, params=params)
            response.raise_for_status()  # Raise an exception for HTTP errors
            data = response.json()
        except requests.exceptions.RequestException as e:
            print(f"Error occurred: {e}")
            break

        # Append the current page of results to the container
        all_works.extend(data.get("results", []))

        # Fetch the next cursor
        cursor = data.get("meta", {}).get("next_cursor", None)
        if not cursor:  # If no next cursor, exit the loop
            print("No more results to fetch.")
            break
    return all_works



def get_works_df_by_query(query):
    cursor = "*"  # Initial cursor (start from the beginning)
    per_page = 200  # Maximum results per page

    # Container to hold all extracted data
    all_works = []

    # Loop to fetch all pages
    while cursor:
        params = {
            "per_page": per_page,  # Limit results per page
            "cursor": cursor,  # Cursor for pagination
            "mailto": email  # Add email for polite pooling
        }

        try:
            # Request data from API
            response = requests.get(base_url, params=params)
            response.raise_for_status()  # Raise an exception for HTTP errors
            data = response.json()
        except requests.exceptions.RequestException as e:
            print(f"Error occurred: {e}")
            break

        # Append the current page of results to the container
        all_works.extend(data.get("results", []))

        # Fetch the next cursor
        cursor = data.get("meta", {}).get("next_cursor", None)
        if not cursor:  # If no next cursor, exit the loop
            print("No more results to fetch.")
            break
    all_works_df = pd.DataFrame(all_works)

    return all_works_df


def reconstruct_abstract(abstract_inverted_index):
    if not abstract_inverted_index:  # Check if the input is None or empty
        return ""  # Return an empty string if no abstract is available

    try:
        # Find the maximum position to determine the size of the resulting list
        max_position = max(pos for positions in abstract_inverted_index.values() for pos in positions)

        # Create a blank list to hold the words at each position
        abstract_tokens = [None] * (max_position + 1)

        # Place each word in its respective positions
        for word, positions in abstract_inverted_index.items():
            for pos in positions:
                abstract_tokens[pos] = word

        # Join the tokens into a single abstract string
        return " ".join(abstract_tokens)

    except Exception as e:
        return ""
```
